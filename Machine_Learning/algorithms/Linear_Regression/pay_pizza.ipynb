```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
d = [[7], [10], [15], [30], [45]]
p = [[8], [11], [16], [38.5], [52]]
```

```python
plt.figure()
plt.xlabel("Diẫmetro (cm)")
plt.ylabel("Preço (R$)")
plt.title("Diâmetro x Preço")
plt.plot(d, p, 'k.')
plt.axis([0,60,0,60])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = [[7], [10], [15], [30], [45]]
y = [[8], [11], [16], [38.5], [52]]
```

```python
modelo = LinearRegression()
```

```python
type(modelo)
```

```python
modelo.fit(x, y)
```

```python
print("Uma pizza de 20 cm de diâmetro deve custar: R$%.2f" % modelo.predict([[2]]))
```

```python
plt.scatter(x, y, color = 'black')
plt.plot(x, modelo.predict(x), color='green', linewidth=2)
plt.xlabel("Diẫmetro (cm)")
plt.ylabel("Preço Preditivo (R$)")
plt.title("Diâmetro x Preço Preditivo")
plt.xticks(())
plt.yticks(())
plt.show()
```
