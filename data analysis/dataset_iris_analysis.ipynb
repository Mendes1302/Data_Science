```python
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
import matplotlib as mat
import pandas as pd
import numpy as np
import time
```

```python
fontsize = 14
ticklabelsize = 14
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
print(len(df))
df.head()
```

```python
kind = iris.target_names
values =  iris.target
```

```python
print(kind)
print(values)
```

```python
df['species'] = pd.Categorical.from_codes(values, kind)
df.head()
```

```python
df['values'] = values
df.head()
```

```python
plt.show()
plt.figure(figsize=(12, 8), dpi=80)
plt.scatter(df['sepal length (cm)'], range(len(df)), c=df['values'])
plt.ylabel('Número da Linha', fontsize=fontsize)
plt.xlabel('Sepal length (cm)', fontsize=fontsize)
plt.title('Gráfico de Dispersão dos Atributos, colorido por marcadores da classe alvo', fontsize=fontsize)
plt.title('Scatter plot of features, colored by target labels', fontsize=fontsize)
```
